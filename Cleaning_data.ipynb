# **#Check the skewness only for numeric columns**

```python
numeric_data = data.select_dtypes(include=np.number)
print(numeric_data.skew())

# Create a box plot for outlier detection - choosing 'price' column instead of 'host_name'
sns.boxplot(x=data["price"])
plt.show()

# Detect outliers using the IQR method - using 'price' for both calculations
Q1 = data["price"].quantile(0.25)
Q3 = data["price"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data["price"] < lower_bound) | (data["price"] > upper_bound)]

# Print the outliers
print(outliers)
```
