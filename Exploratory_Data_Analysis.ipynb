```python
# Calculating mean, median, and mode for the 'calories' column
calories_mean = data['calories'].mean()
calories_median = data['calories'].median()
calories_mode = data['calories'].mode()[0]

print(f"Mean: {calories_mean}")
print(f"Median: {calories_median}")
print(f"Mode: {calories_mode}")
```

```python
# Line plot for 'calories'
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['calories'], marker='o', label='Calories')
plt.title('Line Plot of Calories')
plt.xlabel('Index')
plt.ylabel('Calories')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['category'], data['calories'], color='orange')
plt.title('Calories per Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Calories', fontsize=12)
plt.legend()

plt.xticks(rotation=45, fontsize=10)  # 45-degree tilt
plt.show()
```

```python
# Enhanced bar plot for 'calories' per category
plt.figure(figsize=(14, 8))
sns.barplot( x='category', y='calories', data=data, palette='viridis', edgecolor='black',)
plt.title('Calories per Category', fontsize=18, fontweight='bold')
plt.xlabel('Category', fontsize=14, fontweight='bold')
plt.ylabel('Calories', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, ha='right')
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create the heatmap data
heatmap_data = data.groupby(['category', 'calories'])['calories'].count().unstack(fill_value=0)

# Create the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt='d', linewidths=.5, edgecolor='black')
plt.title('Calories per Category', fontsize=18, fontweight='bold')
plt.xlabel('Calories', fontsize=14, fontweight='bold')
plt.ylabel('Category', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, fontsize=12, ha='right')
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
